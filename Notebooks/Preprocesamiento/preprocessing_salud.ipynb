```python
# Importando librerías de interés
import pandas as pd
import numpy as np
```

```python
#Nombre de columnas del archivo de 2012
cols_salud = [
'entidad',
'munici',
'locali',
'sexo',
'edad',
'aa402ms',
'aa403ms',
'aa404ms',
'localida',
'pondef'
]
```

```python
#Nombre de columnas del archivo de 2016
cols_salud_16 = [
'entidad',
'munici',
'locali',
'sexo',
'edad',
'afv_s',
'afm_s',
'afw_s',
'rural',
'ponde_f'
]
```

```python
# Actividad física de adultos y adolescentes en México en 2012
path = 'actividad_2012.csv'
df_salud_12 = pd.read_csv(path)
df_salud_12 = df_salud_12[cols_salud]
df_salud_12['anio'] = 2012
df_salud_12.head()
```

```python
# Actividad física de adultos y adolescentes en México en 2016
path = 'actividad_2016_15-19.csv'
df_salud_16_adol = pd.read_csv(path, dtype='unicode', encoding='latin1')
df_salud_16_adol = df_salud_16_adol[cols_salud_16]
df_salud_16_adol['anio'] = 2016
df_salud_16_adol['entidad'] = df_salud_16_adol['entidad'].astype(int)
df_salud_16_adol['entidad'] = df_salud_16_adol['entidad'].astype(str)
df_salud_16_adol['munici'] = df_salud_16_adol['munici'].astype(int)
df_salud_16_adol['munici'] = df_salud_16_adol['munici'].astype(str)
df_salud_16_adol['locali'] = df_salud_16_adol['locali'].astype(int)
df_salud_16_adol['locali'] = df_salud_16_adol['locali'].astype(str)

path = 'actividad_2016_mayores.csv'
df_salud_16_mayores = pd.read_csv(path, dtype='unicode', encoding='latin1')
df_salud_16_mayores = df_salud_16_mayores[cols_salud_16]
df_salud_16_mayores['anio'] = 2016
df_salud_16_mayores['entidad'] = df_salud_16_mayores['entidad'].astype(int)
df_salud_16_mayores['entidad'] = df_salud_16_mayores['entidad'].astype(str)
df_salud_16_mayores['munici'] = df_salud_16_mayores['munici'].astype(int)
df_salud_16_mayores['munici'] = df_salud_16_mayores['munici'].astype(str)
df_salud_16_mayores['locali'] = df_salud_16_mayores['locali'].astype(int)
df_salud_16_mayores['locali'] = df_salud_16_mayores['locali'].astype(str)


df_salud_16 = pd.concat([df_salud_16_adol, df_salud_16_mayores], axis=0)
df_salud_16.rename(columns={'afv_s': 'aa402ms',
                   'afm_s': 'aa403ms',
                   'afw_s': 'aa404ms',
                   'rural': 'localida',
                   'ponde_f': 'pondef',
                   },
          inplace=True)
df_salud = pd.concat([df_salud_12, df_salud_16], axis=0)
df_salud.rename(columns={'aa402ms': 'actividad_vigorosa',
                   'aa403ms': 'actividad_moderada',
                   'aa404ms': 'actividad_caminar',
                   'localida': 'urbanizacion',
                   'pondef': 'ponderador',
                   },
          inplace=True)
df_salud.tail()
```

```python
#Creación del ID de localidad concatenando entidad, municipio y número de localidad
df_salud['localidad'] = df_salud['entidad'].astype(str) + '_' + df_salud['munici'].astype(str) + '_' + df_salud['locali'].astype(str)
df_salud.drop(['entidad', 'munici', 'locali'], axis=1, inplace=True)
df_salud.head()
```

```python
#Filtro de valores nulos 
df_salud_filtrada = df_salud[['localidad', 'actividad_vigorosa', 'actividad_moderada', 'actividad_caminar', 'urbanizacion', 'ponderador', 'anio']]
df_salud_filtrada['actividad_vigorosa'] = df_salud_filtrada['actividad_vigorosa'].astype(str).replace(' ','0')
df_salud_filtrada['actividad_moderada'] = df_salud_filtrada['actividad_moderada'].astype(str).replace(' ','0')
df_salud_filtrada['actividad_caminar'] = df_salud_filtrada['actividad_caminar'].astype(str).replace(' ','0')
df_salud_filtrada['urbanizacion'] = df_salud_filtrada['urbanizacion'].astype(str).replace(' ','3')
df_salud_filtrada['ponderador'] = df_salud_filtrada['ponderador'].astype(str).replace(' ','0')
df_salud_filtrada.fillna(0, inplace=True)
df_salud_filtrada['actividad_vigorosa'] = df_salud_filtrada['actividad_vigorosa'].astype(str).replace('nan','0')
df_salud_filtrada['actividad_moderada'] = df_salud_filtrada['actividad_moderada'].astype(str).replace('nan','0')
df_salud_filtrada['actividad_caminar'] = df_salud_filtrada['actividad_caminar'].astype(str).replace('nan','0')
df_salud_filtrada['urbanizacion'] = df_salud_filtrada['urbanizacion'].astype(str).replace('nan','0')

#Pasamos los datos de muestra a población usando la variable de ponderador. Se tiene que multiplicar el ponderador con el valor correspondiente.
df_salud_filtrada['act_vig_tot'] = df_salud_filtrada['actividad_vigorosa'].astype(float) * df_salud_filtrada['ponderador'].astype(float)
df_salud_filtrada['act_mod_tot'] = df_salud_filtrada['actividad_moderada'].astype(float) * df_salud_filtrada['ponderador'].astype(float)
df_salud_filtrada['act_cam_tot'] = df_salud_filtrada['actividad_caminar'].astype(float) * df_salud_filtrada['ponderador'].astype(float)
df_salud_filtrada['ponderador'] = df_salud_filtrada['ponderador'].astype(float)
df_salud_filtrada['localidad'] = df_salud_filtrada['localidad'].astype(str)
df_salud_filtrada.tail(50)
```

```python
#Agrupación por localidad, año, y urbanización. También se saca el promedio de cada una dividiendo entre el ponderador
df_salud_filtrada = df_salud_filtrada.groupby(['localidad',  'anio', 'urbanizacion'], as_index=False).sum().sort_values(by='localidad', ascending=True)
df_salud_filtrada['actividad_vigorosa_promedio'] = df_salud_filtrada['act_vig_tot'].div(df_salud_filtrada['ponderador'])
df_salud_filtrada['actividad_moderada_promedio'] = df_salud_filtrada['act_mod_tot'].div(df_salud_filtrada['ponderador'])
df_salud_filtrada['actividad_caminar_promedio'] = df_salud_filtrada['act_cam_tot'].div(df_salud_filtrada['ponderador'])
df_salud_filtrada.head(10)
```

```python
#Quitamos columnas que no nos importan para la tabla final
df_salud_filtrada = df_salud_filtrada[['localidad', 'urbanizacion', 'anio', 'actividad_vigorosa_promedio', 'actividad_moderada_promedio', 'actividad_caminar_promedio']]
df_salud_filtrada
```

```python
#Exportamos la tabla final
df_salud_filtrada.to_csv('df_salud_filtrada.csv')
```
