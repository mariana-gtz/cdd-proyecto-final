# Limpieza de archivo

## Por enfermedades de pulmón

```python
# Códigos de enfermedades relacionadas al pulmón según INEGI
df_enfermedades_pulmon = ['A150', 'A162', 'C349', 'D381', 'J984']
```

```python
df_filtered = df[df.causa_def.isin(df_enfermedades_pulmon)]
df_filtered.shape
```

```python
# Mapeando nombres de enfermedades con su código
df_filtered.causa_def = df_filtered.causa_def.map(dict_enfermedades)
```

```python
df_filtered.info()
```

```python
# Reinicio de índices post-filtrado
df_filtered.reset_index(inplace=True)
```

## Por lugares de ocurrencia y residencia

```python
# Dando formato a las columnas de localidad, entidad y municipio para poder formar códigos

# Entidades
df_filtered.loc[:, "ent_resid"] = df_filtered["ent_resid"].map('{:02d}'.format)
df_filtered.loc[:, "ent_ocurr"] = df_filtered["ent_ocurr"].map('{:02d}'.format)

# Municipios
df_filtered.loc[:, "mun_resid"] = df_filtered["mun_resid"].map('{:03d}'.format)
df_filtered.loc[:, "mun_ocurr"] = df_filtered["mun_ocurr"].map('{:03d}'.format)

# Localidades
df_filtered.loc[:, "loc_resid"] = df_filtered["loc_resid"].map('{:04d}'.format)
df_filtered.loc[:, "loc_ocurr"] = df_filtered["loc_ocurr"].map('{:04d}'.format)
```

```python
# Creando columna de código de lugar de residencia
df_filtered['lugar_residencia'] = df_filtered['ent_resid'].astype(str) + df_filtered['mun_resid'].astype(str) + df_filtered['loc_resid'].astype(str)
```

```python
# Creando columna de código de lugar de ocurrencia
df_filtered['lugar_ocurrencia'] = df_filtered['ent_ocurr'].astype(str) + df_filtered['mun_ocurr'].astype(str) + df_filtered['loc_ocurr'].astype(str)
```

```python
# Eliminando columnas de entidades, municipios y localidades
df_filtered.drop([
    'ent_resid',
    'mun_resid',
    'loc_resid',
    'ent_ocurr',
    'mun_ocurr',
    'loc_ocurr'
],axis=1, inplace=True)
```

```python
# Mapeando nombres de LUGARES DE RESIDENCIA con su código
df_filtered.lugar_residencia = df_filtered.lugar_residencia.map(dict_lugar)
```

```python
# Mapeando nombres de LUGARES DE OCURRENCIA con su código
df_filtered.lugar_ocurrencia = df_filtered.lugar_ocurrencia.map(dict_lugar)
```

```python
# Eliminando filas con valores nulos
df_filtered = df_filtered.dropna()
```

```python
df_filtered.info()
```

```python
# Eliminando columnas sin lugar de incidencia definido
df_filtered = df_filtered.drop(df_filtered[df_filtered['lugar_residencia'].str.contains('especifica', regex=True)].index)
```

```python
df_filtered.head()
```

```python
# Reinicio de índices post-filtrado
df_filtered.reset_index(inplace=True)
df_filtered[df_filtered.index.duplicated()]
```

## Por tamaño de localidad

```python
# Eliminando aquellas localidades donde el tamaño no está especificado
df_filtered = df_filtered[df_filtered.tloc_ocurr != 99]
```

```python
# Transformando columna de tamaño de localidad de residencia y ocurrencia
df_filtered.tloc_resid = df_filtered.tloc_resid.map(dict_tamanos)
df_filtered.tloc_ocurr = df_filtered.tloc_ocurr.map(dict_tamanos)
```

```python
# Reinicio de índices post-filtrado
df_filtered.drop([
    'index',
    'level_0'
],axis=1, inplace=True)
df_filtered.reset_index(inplace=True)
```

## Por sexo

```python
# Formato a la columna de Sexo
df_filtered.sexo.loc[df_filtered.sexo == 1] = "M"
df_filtered.sexo.loc[df_filtered.sexo == 2] = "F"
df_filtered.sexo.loc[df_filtered.sexo == 9] = "O"
```

```python
df_filtered = df_filtered[df_filtered.sexo != "O"]
```

## Por Edad

```python
# Se limita a edades de un año para arriba, se excluyen valores sin edad definida en años
df_filtered = df_filtered[(df_filtered.edad > 3098) & (df_filtered.edad < 4998) ]
```

```python
# Intercambio códigos por edades
df_filtered.edad = df_filtered.edad.map(dict_edades)
```

```python
# Reinicio de índices post-filtrado
df_filtered.drop([
    'index'
],axis=1, inplace=True)
df_filtered.reset_index(inplace=True)
```

## Por Ocupación

```python
# Filtrando ocupaciones no especificadas
df_filtered = df_filtered[(df_filtered.ocupacion <= 97)]
```

```python
# Intercambio códigos por ocupaciones
df_filtered.ocupacion = df_filtered.ocupacion.map(dict_ocup)
```

```python
# Eliminando filas con valores nulos de ocupación
df_filtered = df_filtered.dropna()
```

## Por Escolaridad

```python
# Filtrando escolaridades no especificadas
df_filtered = df_filtered[(df_filtered.ocupacion != 99)]
```

```python
df_filtered.shape
```

```python
# Intercambio códigos por escolaridades
df_filtered.escolarida = df_filtered.escolarida.map(dict_escolaridad)
```

## Por sitio donde ocurrió la defunción

```python
# Filtrando lugares de defunción no especificadas
df_filtered = df_filtered[(df_filtered.sitio_ocur != 99)]
```

```python
# Intercambio códigos por lugar de la defunción
df_filtered.sitio_ocur = df_filtered.sitio_ocur.map(dict_lugar_defun)
```

## Por Derechohabiencia

```python
# Filtrando derechohabiencias no especificadas
df_filtered = df_filtered[(df_filtered.derechohab != 99)]
```

```python
# Intercambio códigos por tipo de derechohabiencia
df_filtered.derechohab = df_filtered.derechohab.map(dict_derecho)
```

## Por área urbana

```python
# Formato a la columna de Área Urbana
df_filtered.area_ur.loc[df_filtered.area_ur == 1] = "Urbano"
df_filtered.area_ur.loc[df_filtered.area_ur == 2] = "Rural"
df_filtered.area_ur.loc[df_filtered.area_ur == 9] = "No especificado"
```

```python
df_filtered = df_filtered[(df_filtered.area_ur != "No especificado")]
```
